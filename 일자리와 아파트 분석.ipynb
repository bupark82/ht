### 고용통계 데이터 확인

```python
import pandas as pd

job_path = 'C:/Users/jack1/AI2/ht/apt/행정구역_시군구__산업별_고용_20211012173008.xlsx'
job_raw = pd.read_excel(job_path, skiprows=1)
job_data = job_raw[['지역별','전체종사자 (명)']]
job_data.columns = ['지역명','고용자수']
job_data.head()
```

- 고용통계 데이터 지역명을 '시도' '구군' 으로나누기

```python
si_do_list = []
gu_list = []

for i in range(len(job_data)):
    sido = job_data['지역명'][i].split(' ')[0]
    gu = job_data['지역명'][i].split(' ')[1]
    
    if len(sido) != 4:
        si_do_list.append(sido[:2])
    
    else:
        si_do_list.append(sido[0]+sido[2])
        
    
    gu_list.append(gu)

job_data['시도'] = si_do_list
job_data['구군'] = gu_list
job_data
    
```

### 세대수 데이터 확인

```python
house_n_path = 'C:/Users/jack1/AI2/ht/apt/행정구역_시군구_별_주민등록세대수_20211012184431.xlsx'
house_n_raw = pd.read_excel(house_n_path)
house_n_raw.head()
```

  세대별 고용비율은 서울 다음으로 많다는 것을 보여준다.
  
- 경기도 절대적인 고용자수는 많지만 세대별 고용비율은 하위권으로 떨어지는 것을 볼 수 있다.

### 시도별 고용자 수, 세대수를 평균 아파트 매매가격과 확인

```python
def budongsan_price(path):
    row_price = pd.read_excel(path, skiprows=10)
    
    big_col = []
    
    for num, tmp in enumerate(row_price['지 역']):
        if pd.isna(tmp):
            big_col.append(big_col[num-1])
        
        else:
            big_col.append(tmp)
    
    small_col = []
    
    for num in range(len(row_price)):
        tmp_list = list(row_price[['지 역', 'Unnamed: 1', 'Unnamed: 2','Unnamed: 3']].iloc[num])
        
        for tmp in tmp_list[3::-1]:
            if not pd.isna(tmp):
                small_col.append(tmp)
                break
        
        
    row_price.index = [big_col, small_col]
    
    price_t = row_price.drop(['지 역', 'Unnamed: 1', 'Unnamed: 2','Unnamed: 3'], axis=1).T
    
    time_index = []
    for time in price_t.index:
        tmp = time.split(' ')
        time_index.append(tmp[0][:-1]+'.'+tmp[1][:-1])
    
    price_t.index = pd.to_datetime(time_index)
    
    return price_t
            
    
    
```

```python
price_path = 'C:/Users/jack1/AI2/ht/apt/평균매매가격_아파트.xlsx'
price_data = budongsan_price(price_path)
price_data
```

```python
sido_list = []
for i in job_sido.index:
    sido_list.append(price_data.loc['2021-08-01'][i][i])

job_sido['평균매매가격'] = sido_list
job_sido
```

#### 고용자 수, 세대수, 평균 아파트 매매가격 그래프
> 서울과 경기도가 세대수와 고용자수, 평균 매매가격이 가장 높은 것을 알 수 있다.

```python
plt.figure(figsize=(10,6))
plt.scatter(job_sido['고용자수'], job_sido['세대수'], c=job_sido['평균매매가격'], s=job_sido['평균매매가격'] * 0.001,
            cmap='YlOrRd', alpha=0.8)

plt.xlabel('고용자수')
plt.ylabel('세대수')

for name in job_sido.index:
    plt.text(job_sido['고용자수'][name]*1.01, job_sido['세대수'][name]*1.05, name, fontsize=13)
    

plt.colorbar()
plt.ticklabel_format(axis='y', useOffset=False, style='plain')
plt.show()
```

#### 세대수 대비 고용비율, 세대수, 평균 아파트 매매가격 그래프
> - 경기가 평균 아파트 매매 가격은 높은데, 세대수 대비 고용자 비율은 떨어진다.
> - 서울은 역시 평균 아파트 매매 가격, 세대수 대비 고용자 비율도 높다.
> - 울산이 세대수 대비 고용자 비율은 높은데 평균 아파트 가격은 높은편은 아니다.

```python
plt.figure(figsize=(10,6))
plt.scatter(job_sido['세대수대비고용'], job_sido['세대수'], c=job_sido['평균매매가격'], s=job_sido['평균매매가격'] * 0.001,
            cmap='YlOrRd', alpha=0.8)

plt.xlabel('세대수 대비 고용자 비율(%)')
plt.ylabel('세대수')

for name in job_sido.index:
    plt.text(job_sido['세대수대비고용'][name]*1.01, job_sido['세대수'][name]*1.05, name, fontsize=13)
    

plt.colorbar()
plt.ticklabel_format(axis='y', useOffset=False, style='plain')
plt.show()
```

#### 광주의 고용 데이터 확인

```python
local='광주'

job_gugun = job_data[job_data['시도'] == local].groupby('구군').sum()
job_gugun['세대수'] = list(house_n_data[house_n_data['시도'] == local].groupby('구군').sum()['세대수'][:-1])

job_gugun['세대수대비고용'] = job_gugun['고용자수'] / job_gugun['세대수'] * 100
job_gugun['평균매매가격'] = price_data.loc['2021-04-01'][local][job_gugun.index]
job_gugun = job_gugun.dropna()
job_gugun
```

```python
plt.figure(figsize=(12,4))
job_gugun.sort_values(by='고용자수', ascending=False)['고용자수'].plot(kind='bar', color='darkcyan')

plt.axhline(y=job_gugun['고용자수'].mean(), color='orange', linewidth=2, ls='--')
plt.xticks(rotation=45)
plt.xlabel('')
plt.ylabel('고용자수')
plt.show()

plt.figure(figsize=(12,4))
job_gugun.sort_values(by='세대수대비고용', ascending=False)['세대수대비고용'].plot(kind='bar', color='darkcyan')

plt.axhline(y=job_gugun['세대수대비고용'].mean(), color='orange', linewidth=2, ls='--')
plt.xticks(rotation=45)
plt.xlabel('')
plt.ylabel('세대수 대비 고용자수(%)')
plt.show()
```

#### 광주의 고용자 수, 세대수, 평균 아파트 매매가로 그래프 
> 위의 표와 아래 그래프로 확인할 수 있듯이 일자리가 광산구, 북구, 서구에 몰려있다는 것을 알수 있고,
  아파트 매매가격은 남구가 가장 비싸다.

```python
plt.figure(figsize=(10,6))
plt.scatter(job_gugun['고용자수'], job_gugun['세대수'], c=job_gugun['평균매매가격'],
            s = pd.to_numeric(job_gugun['평균매매가격'])*0.001, cmap='YlOrRd', alpha=0.5)

plt.xlabel('고용자수')
plt.ylabel('세대수')

for name in job_gugun.index:
    plt.text(job_gugun['고용자수'][name]*1.01, job_gugun['세대수'][name], name)

plt.colorbar()
plt.ticklabel_format(axis='y', useOffset=False, style='plain')
plt.show()
```

#### 광주의 고용비율, 세대수, 평균 아파트 매매가로 그래프 그리기
> * 북구는 거주하는 세대수가 가장 많지만 세대수 대비 고용비율도 낮고 아파트 평균 가격도 매우 낮게 측정되어 있다.
> * 이외 북구, 광산구, 서구, 동구 순으로 고용자비율과 아파트 가격이 비례했다.
> * 남구는 특수한 케이스로 아파트 가격이 매우 비싸면서 고용자비율이 적었는데, 이는 남구의 특이성에 기인한다.
> 남구는 초기에 불편한 교통과 편이성으로 인해 사용하기 힘들지만 큰 부지가 존재 하였고, 
건설사가 큰 평수와 주변의 환경(무등산 등)을 이용하여 고급 아파트 전략을 사용하였다.
그 결과 고급 아파트의 비율이 높아져 평균 가격이 높게 측정되었다.

특이점으로 인한 남구를 제외하면 아파트 가격과 일자리가 어느정도 비례하는걸 알수 있다.

```python
plt.figure(figsize=(10,6))
plt.scatter(job_gugun['세대수대비고용'], job_gugun['세대수'], c=job_gugun['평균매매가격'],
            s = pd.to_numeric(job_gugun['평균매매가격'])*0.001, cmap='YlOrRd', alpha=0.5)

plt.xlabel('세대수 대비 고용자수(%)')
plt.ylabel('세대수')

for name in job_gugun.index:
    plt.text(job_gugun['세대수대비고용'][name]*1.01, job_gugun['세대수'][name], name)
    
plt.colorbar()
plt.ticklabel_format(axis='y', useOffset=False, style='plain')
plt.show()
```
